## Visualize Router Data

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from tools import *
from localizer import *
```

```python
lab_loc = localizer()
lab_loc.load_profile('lab_profile')

X, y = load_data_from_folder('../datasets', lab_loc.profile, keep_percent=1.0)
# get only the RSSI, not interested in the Quality for the moment
X = X[:, ::2]
print('X shape: ', X.shape)
```

*Display a graph of wifi signal over time for a given router*

```python
routers = X.T
plt.figure(figsize=(16, 9))
for router in routers:
    plt.plot(router)

plt.xlabel('time (s)')
plt.ylabel('RSSI (dB)')
plt.grid(True)
plt.show()
```

*Find standard deviation for each router in each location*

```python
split = np.array(np.split(routers, 9, axis=1))
split = np.reshape(split, (20, 9, 100))
print(split.shape)
```

```python
plt.plot(split[0,0])
plt.show()
```

```python
stds = np.std(split, axis=2)
print(stds.shape)
print(np.amax(stds))
max_idx = np.unravel_index(np.argmax(stds), stds.shape)
print(stds[max_idx])
```

```python
plt.plot(split[max_idx])
```

```python
for loc in split[19]:
    plt.plot(loc)
plt.show()
```

## Normalize the data

```python
normalized = split
normalized -= np.mean(normalized)
normalized /= np.std(normalized, axis=0)
```

```python
for cell in normalized:
    plt.plot(cell[0])
plt.show()
```
